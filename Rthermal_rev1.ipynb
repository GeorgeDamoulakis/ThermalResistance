```python
import pandas as pd
import numpy as np
```

Import the CSV

```python
data = pd.read_csv (r'C://Users//US02244//python_files//Rthermal//p1.csv')
data
```

You create the column Waste Heat

```python
data['WasteHeat_W'] = (data['Current (A)'] * data['Voltage (V)']) - data['Optical power (W)']
data
```

delete the first line

```python
df = data.iloc[1: , :]
df
```

Find the values of intercept and Cold WL

```python
y = df['Wavelength (nm)']
x = df['WasteHeat_W']
slope_intercept = np.polyfit(x,y,1)
intercept = slope_intercept[1]
ColdWL_cst = 0.344088810685303
#print(intercept, ColdWL_cst)
```

Create two columns for intercept and Cold WL

```python
df.loc[:,'intercept'] = intercept
df.loc[:,'ColdWL_cst'] = ColdWL_cst
df
```

Create the column Temperature Change

```python
df.loc[:,'Temp_Change_C'] = ( df['Wavelength (nm)'] - df['intercept'] ) / df['ColdWL_cst'] 
df
```

Find the Theremal resistance

```python
y1 = df['Temp_Change_C']
x1 = df['WasteHeat_W']
slope_intercept_1 = np.polyfit(x1,y1,1)
Theramal_resistance = slope_intercept_1[0]

print('Theramal Resistance =', round(Theramal_resistance, 2) , ' degC / W') 
```
